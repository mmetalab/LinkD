# Interactive Web Server for LLM Planning Agent

This notebook demonstrates how to create an interactive web server for the LLM Planning Agent.
The web server allows users to:
- Enter natural language queries
- See the generated analysis plan
- Watch plan execution in real-time
- View detailed results and summaries
- Click example queries to automatically populate the query field

We'll use **Gradio** for the web interface as it provides excellent interactivity and is easy to deploy.

## 1. Install Required Packages

```python
import sys
sys.path.insert(0, '..')
import os
os.chdir('..')

try:
    from dotenv import load_dotenv
    load_dotenv('../.env')
except ImportError:
    pass

# Install gradio if not already installed
# !pip install gradio

import sys
import subprocess

def install_package(package):
    try:
        __import__(package)
        print(f"✓ {package} is already installed")
    except ImportError:
        print(f"Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"✓ {package} installed successfully")

# Check/install gradio
install_package("gradio")

print("\nAll packages ready!")
```
